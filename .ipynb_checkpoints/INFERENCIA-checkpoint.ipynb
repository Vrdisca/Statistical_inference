```python
df2.loc[df2['PHQ_GRP'] < 5.1, 'Php_grp2'] = 'Sintomas lev/sem'  
df2.loc[((df2['PHQ_GRP'] >= 5.1) & (df2['PHQ_GRP'] < 10.2)), 'Php_grp2'] = 'Sintomas lev/sem'  
df2.loc[((df2['PHQ_GRP'] >= 10.2) & (df2['PHQ_GRP'] < 15.3)), 'Php_grp2'] = 'sintomas mod/sev'
df2.loc[((df2['PHQ_GRP'] >=15.3) & (df2['PHQ_GRP'] < 19)), 'Php_grp2'] = 'sintomas mod/sev'
df2.loc[(df2['PHQ_GRP'] >= 19), 'Php_grp2'] = 'sintomas mod/sev'
```

```python
df2.head()
```

```python
import scipy 
from scipy import stats as st
```

```python
#TESTE 1: ASSOCIAÇÃO ENTRE ADHERENCE E SINTOMAS:
# H0= não há; H1= Há

#TABELA DE CONTINGÊNCIA :
Cross = pd.crosstab(df2.ADHERENCE,df2.Php_grp2, margins=False)
Cross
```

```python
stat, p, dof, expected = st.chi2_contingency(Cross)
```

```python
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
# TESTE 2: Relação entre Php_grp e HEIscore
gropo_ameno = df2.query('Php_grp2 == "Sintomas lev/sem"').HEI2015_TOTAL_SCORE
gropo_risco = df2.query('Php_grp2 == "sintomas mod/sev"').HEI2015_TOTAL_SCORE
```

```python
# teste de normalidade
stat_test, p_valor = st.shapiro(gropo_ameno)
print('P-valor: ', round(p_valor,2), ', nível de significancia (alfa):',  round(0.05,2))
```

```python
stat_test, p_valor = st.shapiro(gropo_risco)
print('P-valor: ', round(p_valor,2), ', nível de significancia (alfa):',  round(0.05,2))
```

```python
#Teste de homogeinedade das variancias (variancias iguais)
#H0 = amostras possuem variâncias iguais; H1: amostras possuem variâncias diferentes

stat_test, p_valor = st.levene(gropo_ameno, gropo_risco, center='mean')
print('P-valor: ', round(p_valor,2), ', nível de significancia (alfa):',  round(0.05,2))
#as amostras possuem variâncias iguais
```

```python
#Teste de hipotese independente
# H1 = médias diferentes nos grupos na população; H0 = médias iguais
stat_test, p_valor = st.ttest_ind(a=gropo_ameno, b=gropo_risco, equal_var=True , alternative='two-sided')
print('P-valor: ', round(p_valor,2), ', nível de significancia (alfa):',  round(0.05,2))
#medias são diferentes---> não descarta hipotese nula
```
